### 1) Create a vector, scalar, matrix and tensor with values of your choosing using tf.constant()

```python
import tensorflow as tf
Vector = tf.constant(2)
print("Dimension of Vector is : ",Vector.ndim)
Scalar = tf.constant([2,3])
print("Dimension of Scalar is : ",Scalar.ndim)
Matrix = tf.constant([[1,2],
                     [3,4]])
print("Dimension of Matrix is : ",Matrix.ndim)
Tensor = tf.constant(tf.zeros(shape=(2,3,5)))
print("Dimension of Tensor is : ",Tensor.ndim)
```

### 2) Find the shape, rank and size of the tensors you created in 1

```python
print("Shape of the Tensor is :",Tensor.shape)
print("Rank of the Tensor is :",Tensor.ndim)
print("Size of the Tensor is :",tf.size(Tensor)) 
```

### 3) Create two tensors containing random values between 0 and 1 with shape [5, 300]

```python
tf.random.set_seed(42)
Tensor_1 = tf.constant(tf.random.uniform(shape=[5,300]))
Tensor_2 = tf.constant(tf.random.uniform(shape=[5,300]))
Tensor_1 , Tensor_2
```

### 4) Multiply the two tensors you created in 3 using matrix multiplication

```python
tf.matmul(Tensor_1,Tensor_2,transpose_b=True)
```

### 5) Multiply the two tensors you created in 3 using dot product

```python
tf.tensordot(Tensor_1,Tensor_2,axes=0)
```

### 6) Create a tensor with random values between 0 and 1 with shape [224, 224, 3]

```python
Tensor_3 = tf.random.uniform(shape=(224,224,3))
Tensor_3
```

### 7) Find the min and max values of the tensor you created in 6

```python
Min_value_of_Tensor_3 = tf.math.reduce_min(Tensor_3).numpy()
Max_value_of_Tensor_3 = tf.math.reduce_max(Tensor_3).numpy()
Min_value_of_Tensor_3 , Max_value_of_Tensor_3
```

### 8) Created a tensor with random values of shape [1, 224, 224, 3] then squeeze it to change the shape to [224, 224, 3]

```python
Tensor_4 = tf.random.uniform(shape=(1,224,224,3))
squeezed_tensor4 = tf.squeeze(Tensor_4)
squeezed_tensor4.shape
```
